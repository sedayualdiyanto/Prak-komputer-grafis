```python
import numpy as np
import matplotlib.pyplot as plt

plt.grid(True)
plt.axis('on')

plt.axis([-16,16,16,-16])
plt.axes().set_aspect('equal')
#custom grid
x1=-16
x2=16
y1=16
y2=-16

dx=.7
dy=-.7
for x in np.arange(x1,x2,dx):
  for y in np.arange(y1,y2,dy):
    plt.scatter(x,y,s=1,color='yellow')

#square box
plt.plot([-7,7],[-7,-7],linewidth=2,color='k')
plt.plot([7,7],[-7,7],linewidth=2,color='k')
plt.plot([7,-7],[7,7],linewidth=2,color='k')
plt.plot([-7,-7],[7,-7],linewidth=2,color='k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([0,150,100,0])
r=40
alpha1=np.radians(0)
alpha2=np.radians(360)
dalpha=np.radians(2)
xc=75
yc=50
plt.scatter(xc,yc,s=10,color='k')
for alpha in np.arange(alpha1,alpha2,dalpha):
  x=xc+r*np.cos(alpha)
  y=yc+r*np.sin(alpha)
  plt.scatter(x,y,s=5,color='k')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Mengatur batas sumbu
plt.axis([-10, 140, 90, -10])

# Menyalakan sumbu
plt.axis('on')

# Menampilkan grid
plt.grid(True)

# Menambahkan judul
plt.title('Contoh Sumbu')

# Menentukan langkah grid
dx = 5
dy = -5

# Menggambar garis grid
for x in np.arange(-10, 140, dx):
    for y in np.arange(90, -10, dy):
        plt.scatter(x, y, s=1, color='lightgray')

# Menggambar sumbu koordinat
plt.arrow(0, 0, 20, 0, head_length=4, head_width=3, color='k')
plt.arrow(0, 0, 0, 20, head_length=4, head_width=3, color='k')

# MenampilkanÂ plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set batas sumbu
plt.axis([-10, 148, 90, -10])

# Nyalakan sumbu
plt.axis('on')

# Matikan grid
plt.grid(False)

# Judul plot
plt.title("Text Samples")

# Contoh teks dengan ukuran berbeda
plt.text(20, 10, 'small text', fontsize="small")
plt.text(20, 15, 'normal text')
plt.text(20, 20, 'large text', fontsize="large")

# Contoh teks tebal dan miring
plt.text(20, 30, 'large bold text', fontsize="large", fontweight='bold')
plt.text(20, 35, 'large bold, italic text', fontsize="large", fontweight='bold', style='italic')
plt.text(20, 40, 'large, pure, bold italic text', fontsize="large", fontweight='bold', style='italic', color=(.5, 0, .5))
plt.text(20, 45, 'large, light purple, bold italic text', fontsize="large", fontweight='bold', style='italic', color=(.8, 0, .8))

# Contoh teks berwarna
plt.text(20, 50, 'light purple text', color=(.8, 0, .8))

# Contoh teks dengan rotasi
plt.text(100, 50, 'text at 45 degrees', rotation=45, color='k')
plt.text(100, 40, 'text at 60 degrees', rotation=-60, color='g')

# Contoh teks dengan rumus matematika
plt.text(28, 65, r'$P(\lambda)-2 \pi c^{2} h\int_{\lambda1}^{\lambda2}\frac{\lambda^{-5}\epsilon} {e^{\frac{hc}{\lambda k t}}-1}d\lambda$', fontsize="large")

# TampilkanÂ plot
plt.show()
```
